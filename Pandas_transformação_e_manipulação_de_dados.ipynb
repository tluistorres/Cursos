# Mtodo groupby( )

Passaremos o comando dados.groupby( ) numa nova clula. Nos parnteses, informaremos o nome da coluna a partir da qual desejamos agrupar os dados do nosso dataframe: 'Tipo'. Ento:

```python
dados.groupby('Tipo')
```

O mtodo groupby( ) funciona em conjunto com outras funes do Pandas, como a sum de soma ou a mean que  a mdia. Com essas funes, podemos fazer agregaes conforme o agrupamento que realizamos.

Aplicaremos a funo de mdia, afinal,  isso que queremos calcular quando o agrupamento for feito. Aps o groupby( ), vamos passar .mean( ).

```python
dados.groupby('Tipo').mean()

# Ocorrer erro.
```

Nesse ponto,  importante nos atentarmos a um detalhe. Analisando rapidamente as colunas do nosso dataframe, notaremos que temos tanto colunas numricas quanto colunas de texto. No cdigo acima, fizemos um agrupamento e aplicamos um mtodo de agregao, o mean( ). .

Se voc estiver usando a verso 1 da biblioteca Pandas, automaticamente esse mtodo de agregao ser aplicado unicamente nas colunas numricas. No entanto, se voc estiver utilizando da verso 2 em diante, esse comportamento no ser o mesmo. Nesse caso, ao fazer o agrupamento e aplicar o mtodo de agregao, ele ser aplicado a todas as colunas, incluindo as de texto. Isso pode gerar um erro.

Para evitar que isso acontea, podemos especificar um parmetro no mtodo mean( ) (e nos outros mtodos de agregao tambm): numeric_only=True.

Com isso, especificamos que a mdia ser calculada apenas nas colunas numricas e no teremos nenhum erro. Ento:

```python
dados.groupby('Tipo').mean(numeric_only=True)
```

Ao executar essa clula, estamos agrupando nosso dataframe dados conforme a coluna Tipo e calculando a mdia de valores conforme esse agrupamento.

O retorno  um dataframe onde os ndices so os tipos de imveis presentes na nossa base de dados, porque agrupamos nossos dados a partir dessa coluna.

# Para saber mais: entendendo o groupby

O mtodo groupby do Pandas permite agrupar e resumir dados de um DataFrame, com base em um ou mais critrios. Esses critrios podem ser variveis numricas ou categricas, como colunas ou ndices do DataFrame.

A ideia por trs do groupby  dividir os dados em grupos com base nos critrios selecionados e, em seguida, aplicar uma operao a esses grupos. Essa operao pode ser uma funo de agregao, como soma, mdia, contagem, desvio padro, entre outras, ou mesmo uma operao personalizada definida pela pessoa usuria.

Esse mtodo possui diversos parmetros, alguns deles so:

  1. by: esse  o parmetro mais comum e  usado para especificar a coluna ou colunas que queremos agrupar. Como argumento dele, podemos passar o nome de uma coluna ou uma lista de nomes de colunas;

  2. axis: utilizamos esse parmetro para especificar o eixo ao longo do qual queremos agrupar. O valor padro dele  0, o que significa que as linhas sero agrupadas. Se quisermos agrupar as colunas, devemos definir esse parmetro como 1;

  3. sort: esse parmetro  um booleano (True ou False) que indica se os grupos devem ser ordenados pelo valor da coluna de agrupamento. O valor padro  True;

  4. dropna: utilizamos esse parmetro para controlar se os valores ausentes (NaN) sero excludos ou no durante o processo de agrupamento. O valor padro  True.


Vamos analisar um exemplo simples utilizando o groupby.

Suponha que temos o seguinte DataFrame:

Cdigo do DataFrame:


```python
import pandas as pd

df = pd.DataFrame({
   'Animal': ['Cachorro', 'Gato', 'Elefante', 'Cachorro', 'Gato', 'Elefante'],
   'Cor': ['Preto', 'Branco', 'Cinza', 'Marrom', 'Preto', 'Marrom'],
   'Quantidade': [2, 3, 1, 4, 2, 2]
})
df
```

Aqui ns temos um DataFrame com dados referentes a alguns animais. Vamos utilizar o groupby para agrupar os dados pelo tipo de animal e realizarmos a soma da coluna "Quantidade" para conferirmos o total de cada tipo de animal:

```python
df.groupby('Animal').sum(numeric_only=True)
```

Podemos tambm usar o mtodo groupby para agrupar os dados por mltiplas colunas. Por exemplo, vamos agrupar pelo tipo do animal e tambm pela cor e realizar a soma da coluna "Quantidade":

```python
df.groupby(['Animal', 'Cor'])[['Quantidade']].sum()
```

Assim, podemos observar que o groupby  um mtodo muito til para situaes em que precisamos dividir os dados em grupos para analisar o comportamento desses grupos. Utilizando esse mtodo, ns podemos responder perguntas especficas sobre nossos conjuntos de dados.

# Para saber mais: lidando com dados nulos

Tratar dados nulos em uma base de dados  muito importante porque esses dados faltantes podem interferir diretamente na qualidade das anlises e nos resultados obtidos. Quando existem dados nulos em um conjunto de dados, podem ocorrer problemas como vis, reduo da acurcia e impreciso nas informaes obtidas.

Alm disso, algumas anlises e modelos estatsticos no so capazes de lidar com a presena de dados faltantes. Por isso, o tratamento dos dados nulos  fundamental para garantir a confiabilidade e a preciso das anlises e decises tomadas a partir desses dados.

Existem diversas formas de tratar dados nulos com o Pandas. Algumas das principais formas so:

 1. Remover os dados nulos:  possvel remover as linhas ou colunas que possuem valores nulos utilizando o mtodo dropna( ) . Esse mtodo remove todas as linhas ou colunas que possuem pelo menos um valor nulo.

 2. Preencher os dados nulos: Utilizando o mtodo fillna( ), podemos preencher os valores nulos com um valor especfico. Alm disso, tambm  possvel utilizar argumentos especficos do mtodo fillna( ) como o method=ffill ou method=bfill para preencher os valores nulos com o valor anterior ou posterior, respectivamente.

 3. Interpolar os dados nulos:  possvel utilizar o mtodo interpolate( ) para preencher os valores nulos com valores interpolados, ou seja, valores calculados a partir dos valores vizinhos.

 importante escolher a melhor forma de tratar os dados nulos de acordo com a situao especfica e o objetivo da anlise de dados.

# Para saber mais: preciso de valores numricos

Em geral, quando os dados so muito grandes e temos pouca memria disponvel,  comum utilizar tipos de dados mais compactos para reduzir o consumo de memria. Mas,  sempre importante garantir que a escolha do tipo de dados no prejudique a preciso ou a acurcia dos resultados.

Quando trabalhamos com nmeros inteiros com Python, podemos ter diversos tipos de dados, cada um com suas limitaes e caractersticas. Durante as aulas, trabalhamos com o int64, um inteiro com preciso de 64 bits. Para entender o significado dessa preciso,  importante conhecer alguns termos tcnicos, como byte e bit:

 - Byte:  uma unidade de medida de informao, que representa um  conjunto de 8 bits.

 - Bit:  a menor unidade de informao utilizada em sistemas digitais, podendo assumir os valores de 0 ou 1.

# Tipo Inteiro

Com os conceitos de bit e byte claros, podemos, ento, entender melhor o significado do int64, que  o tipo inteiro que utiliza 8 bytes de armazenamento - 8 bits em cada byte, resultando em 64 bits ao total. Esse tipo inteiro  capaz de representar nmeros muito grandes, que podem variar de -9.223.372.036.854.775.808 a 9.223.372.036.854.775.807.

Alm dele, temos outros inteiros que podem ter sua preciso definida, como o int32, um tipo de dado inteiro que utiliza 4 bytes - 8 bits em cada byte, resultando em 32 bits ao total. Ele  capaz de representar nmeros inteiros menores do que os representados pelo int64, com um mximo de -2.147.483.648 a 2.147.483.647.

Pode ser mais comum encontrar os tipos int64 e int32, mas pode ser necessrio, em algumas situaes, utilizar outros tipos de dados inteiros, como o int8 ou o int16. Tipos de dados como esses so teis quando  preciso economizar mais memria e no se est trabalhando com grandes valores. Voc pode conferir os tipos de inteiros na tabela abaixo:

A escolha de qual valor de preciso vai depender da situao e da natureza dos dados sendo manipulados. Se os valores que esto sendo analisados so relativamente pequenos, o uso do int32, por exemplo, pode ser o suficiente, podendo economizar espao em memria. # Para saber mais: o que  HTML?

HTML (Hypertext Markup Language - Linguagem de marcao de hipertexto)  o cdigo que usamos para estruturar uma pgina web e seu contedo. O HTML consiste em uma srie de elementos que so usados para delimitar ou agrupar diferentes partes do contedo para que ele aparea ou atue de determinada maneira.

Um documento HTML comea com a tag <"html"> e termina com a tag <"/html">.

Ele tambm contm um cabealho (<"head">) que descreve o ttulo da pgina e outras informaes relevantes; e um corpo (<"body">) que contm o contedo da pgina. O cabealho e o corpo so delimitados pelas tags <"head"> e <"/head">, e <"body"> e <"/body">, respectivamente.

Dentro do corpo podemos usar vrios elementos para criar a estrutura do contedo. Note cada detalhe no exemplo a seguir:

tags <"h1"> a <"h6"> : criar ttulos e subttulos.

tags <"p">: criar pargrafos.

tags <"ul"> e <"li"> : criar listas no ordenadas.

tags <"img">: inserir imagens.

tags <"table">, <"tr">, <"th"> e <"td">: criar tabelas.

Alm disso, podemos usar atributos nos elementos HTML para especificar propriedades adicionais.

Por exemplo: podemos usar o atributo src na tag <img> para indicar o endereo da imagem que queremos mostrar, ou o atributo href na tag <a> para indicar o endereo do link que queremos criar.

# Para saber mais: aprofundando na normalizao

Normalizao de dados  um processo importante em cincia de dados que tem como objetivo organizar e padronizar dados para facilitar a anlise e comparao entre eles. Quando se trata de dados no formato JSON (JavaScript Object Notation - Notao de Objetos JavaScript),  comum que eles estejam aninhados, o que pode dificultar sua anlise e manipulao.


A biblioteca Pandas possui uma funo chamada json_normalize() que permite a transformao dos dados em um formato tabular, facilitando a visualizao e anlise das informaes. A seguir vamos aprender como usar essa funo para normalizar diferentes tipos de JSON em DataFrames.


# Normalizando um JSON simples

Abaixo temos uma varivel chamada dados e dentro dela h um objeto JSON com trs chaves e seus respectivos valores:




```python

dados = {'Pesquisa': 'Principais Indicadores de Doenca Cardiaca', 'Ano': 2020, 'Numero_Pacientes':3}

# Para normalizar essa varivel, podemos passa-l dentro da funo json_normalize e analisar o DataFrame obtido:

df = pd.json_normalize(dados)
df
```

O resultado parece timo! Temos 3 colunas que so as nossas chaves e uma linha contendo os valores para cada chave.

# Normalizando um JSON com vrios nveis

Existem situaes em que o arquivo JSON pode conter mais de um objeto, como  o caso do exemplo abaixo, no qual temos uma lista, armazenada na varivel json_lista, contendo dois objetos JSON:

```python
json_lista = [
    { 'ID': '01', 'Faixa_etaria': '55-59', 'Sexo_biologico': 'feminino'},
    { 'ID': '02', 'Faixa_etaria': '80 ou +', 'Sexo_biologico': 'feminino'}
]

# Para normalizar essa lista podemos aplicar a funo json_normalize:

pd.json_normalize(json_lista)
```

A funo json_normalize()  capaz de converter cada registro da lista em uma linha de forma tabular.

# Normalizando um JSON com uma lista aninhada

Vamos comear analisando a normalizao quando os dados so um dicionrio. Temos um dicionrio armazenado na varivel json_obj. Note que na chave Saude temos outro dicionrio:


```python
json_obj = {
    'ID': '01',
    'Faixa_etaria': '55-59',
    'Sexo_biologico': 'Feminino',
    'Saude': {'Dificuldade_caminhar': 'Nao',
              'Atividade_fisica': 'Sim',
              'IMC': 16.6,
              'Doenca_cardiaca': 'Nao',
          }
      }

# Fazendo a normalizao:

pd.json_normalize(json_obj)

```

O DataFrame gerado possui uma coluna para cada informao contida no dicionrio que estava em Saude. As colunas criadas possuem o prefixo Saude., pois as informaes vieram dessa mesma chave.

Tambm podemos fazer isso utilizando o parmetro record_path como ['Pacientes']. Esse parmetro  usado na funo *pd.json_normalize()* para especificar o caminho para os registros que devem ser normalizados em um DataFrame separado:

Importante: Na aula ns realizamos a normalizao em um arquivo com o formato JSON. No entanto, a funo json_normalize() aceita apenas um dicionrio ou uma lista de dicionrios. # Importando a biblioteca Pandas

```python
# Importando a biblioteca Pandas

import pandas as pd

# Importando o mdulo JSON

import json

# Lendo o arquivo json usando o mdulo Python JSON

with open('pacientes_2.json','r') as f:
    dados = json.loads(f.read())

# Normalizando os dados com os parmetros record_path e meta

pd.json_normalize(dados, record_path='Pacientes', meta=['Pesquisa', 'Ano'])
```

E, assim, aprendemos como normalizar arquivos JSON simples, com mltiplos nveis e aninhados.

# Para saber mais: obtendo arquivos JSON de APIs

Existem vrios formatos que podem ser usados em APIs e um dos mais comuns  o JSON (JavaScript Object Notation). O JSON  um formato leve de troca de informaes entre sistemas que utiliza uma notao baseada em objetos do JavaScript.

Ns temos a seguinte situao-problema: Uma empresa de tecnologia est em fase de expanso e deseja otimizar sua base de contatos e a forma como gerencia suas interaes com os clientes e parceiros.

A equipe de vendas e atendimento ao cliente est enfrentando dificuldades para acompanhar todos os contatos, entender as necessidades dos diferentes perfis de clientes, e personalizar o atendimento de forma eficiente. Para solucionar isso, a empresa precisa de uma maneira organizada e automatizada para coletar, gerenciar e segmentar informaes de seus clientes e parceiros, como nome, e-mail, endereo, telefone e empresa em que trabalham.

Nesta situao, podemos acessar a API JSONPlaceholder para coletar os dados dos contatos, que incluem nome, e-mail, endereo, telefone, entre outros.

Com esses dados, a equipe pode melhorar suas operaes e a comunicao com os clientes.

Uma maneira comum de acessar APIs  por meio da biblioteca requests. Para isso, tambm  necessrio importar o mdulo json:

```python
import requests

import json

# Os dados da JSONPlaceholder sero obtidos com a funo request.get('url'). Dentro dela ns passamos o endereo da API:

dados_usuarios = requests.get('https://jsonplaceholder.typicode.com/users')

# Conseguimos recuperar os resultados usando a funo json.loads()
# Para isso,  necessrio passar a varivel dados_usuarios, criada anteriormente com o atributo text que retorna o contedo da resposta.

resultado = json.loads(dados_usuarios.text)

# Em seguida, podemos visualizar o DataFrame.

pd.DataFrame(resultado)
```

Agora sabemos como utilizar a biblioteca requests e o mdulo JSON para obter dados de APIs!

# xls

O xls  um formato de arquivo do Excel mais antigo e usado at a verso de 2003.

# xlsx

J o xlsx  o formato de arquivo padro do Excel a partir da verso 2007. Esse formato  baseado em XML (Extensible Markup Language - Linguagem de marcao extensvel) e  amplamente suportado por outras ferramentas de planilha eletrnica, incluindo o Google Planilhas.

Ateno: Se surgiram algumas dvidas com o termo XML, no se preocupe! Mais adiante teremos uma aula sobre esse tema. Nela voc entender o que  esse formato, na medida em que l e escreve em XML.

# xlsm

H tambm o xlsm, uma extenso de arquivo usada pelo Excel para armazenar planilhas que contm as macros. Estas so sequncias de comandos ou instrues que podem ser executadas automaticamente para realizar tarefas especficas na planilha.

Ento, o formato xlsm permite que as macros sejam salvas juntamente com a planilha, para que possam ser executadas sempre que a planilha for aberta.

# xlsb

Por fim, temos o formato xlsb, uma extenso de arquivo usada pelo Excel para armazenar planilhas em formato binrio. A codificao binria permite que as planilhas sejam abertas e salvas com mais rapidez do que aquelas no formato xlsx.

J os formatos odf, ods e odt so arquivos abertos, livres e universais que podem ser usados por qualquer software, ou seja, foram criados para serem independentes de plataformas, isso significa que podem ser usados em sistemas operacionais diferentes, como Windows, Linux e Mac OS.

Alm disso, so independentes de aplicativos e podem ser usados em vrios programas diferentes, incluindo o OpenOffice, o LibreOffice, o Google Docs e o Microsoft Office.

# Mas o que isso quer dizer? O que  esse tal de UTF-8?

O erro de encoding ocorre quando a biblioteca Pandas no consegue interpretar corretamente os caracteres de um arquivo CSV. Isso pode acontecer quando ele contm caracteres especiais que no so reconhecidos pela biblioteca Pandas ou quando foi salvo em um formato de codificao diferente do esperado.

Para resolver esse erro,  necessrio identificar a codificao correta do arquivo CSV e especific-la ao carregar o arquivo com a biblioteca Pandas. Essa codificao padro  o UTF-8, mas em alguns casos, o arquivo pode ter sido salvo com uma codificao diferente, como ISO-8859-1.

O UTF-8  uma codificao de caracteres universal usada para representar caracteres de diferentes idiomas de forma compatvel com a internet e com sistemas de computador em geral. A sigla UTF significa Unicode Transformation Format (Formato de Transformao Unicode) e o nmero 8 indica que essa codificao associa uma sequncia de 1 a 4 bytes (8 a 32 bits) com cada caractere.

A codificao UTF-8  amplamente utilizada na internet e em sistemas de computador em todo o mundo, pois permite a representao de caracteres de diferentes idiomas em um nico conjunto de caracteres. Alm disso, essa codificao  capaz de preservar a compatibilidade com outras mais antigas, como ASCII, o que a torna uma escolha popular para a criao e compartilhamento de arquivos de texto.

Pode ser que voc esteja pensando neste momento: como ser que a Giovanna conseguiu descobrir qual  a codificao do arquivo que ela est tentando ler?

Existem algumas formas para descobrir isso, no entanto, teremos a oportunidade de experimentar uma maneira prtica de fazer isso no prprio Google Colab.

Ns podemos usar uma biblioteca chamada chardet para detectar o encoding de um arquivo CSV. Para utilizar essa biblioteca no Google Colab, basta realizar sua importao:

```python
import chardet

# Em seguida, digitamos o seguinte bloco de cdigo:

with open('/content/dados.csv', 'rb') as file:
    print(chardet.detect(file.read()))
```

Note que a primeira linha deste cdigo abre o arquivo CSV em modo de leitura binria rb e atribui o objeto de arquivo retornado a uma varivel chamada file.

Na segunda linha do cdigo, o contedo do arquivo CSV  lido usando o mtodo read() e o resultado  passado para a funo chardet.detect(), que retorna um dicionrio contendo informaes sobre a codificao mais provvel do arquivo. O resultado  impresso com a funo print() que mostra a codificao e a confiana associada a essa codificao.

# Para saber mais: clusulas SQL

SQL (Structured Query Language - Linguagem de consulta estruturada)  uma linguagem de consulta utilizada em bancos de dados relacionais para inserir, atualizar, consultar e gerenciar dados.

Para fazer tudo isso,existem as clusulas SQL, que so componentes fundamentais das instrues SQL, permitindo especificar detalhes sobre como a consulta ou a operao de banco de dados devem ser executadas. As clusulas so usadas para filtrar, classificar, agrupar e limitar os resultados das consultas.

As instrues SQL podem ser compostas por uma ou mais clusulas que fornecem informaes adicionais sobre o que a consulta deve fazer. As clusulas mais comuns so:

 - SELECT: especifica quais colunas devem ser selecionadas na consulta.

 - FROM: especifica as tabelas do banco de dados que devem ser consultadas.

 - WHERE: filtra os resultados da consulta com base em uma ou mais condies especificadas.

 - ORDER BY: classifica os resultados da consulta em ordem crescente ou decrescente com base em uma ou mais colunas.

 - GROUP BY: agrupa os resultados da consulta com base em uma ou mais colunas.

 - LIMIT: limita o nmero de linhas retornadas pelos resultados da consulta.

Temos um exemplo simples de sintaxe SQL que usa o comando SELECT para consultar dados de uma tabela de colaboradores(as) em um banco de dados:

```python
SELECT nome, sobrenome, salario

FROM colaboradores

WHERE departamento = 'vendas'
```

Nessa instruo SQL, a clusula SELECT  usada para especificar as colunas que se deseja consultar da tabela "colaboradores", incluindo "nome", "sobrenome" e "salario". A clusula FROM  usada para especificar a tabela que se deseja consultar, que  "colaboradores" neste caso.

A clusula WHERE  usada para filtrar os resultados da consulta com base em uma condio especificada. Neste exemplo, a condio  departamento = 'vendas', o que significa que a consulta retornar apenas as pessoas colaboradoras que trabalham no departamento de vendas.

 1. Definio da consulta SQL: Primeiro, definimos o que queremos fazer com o banco de dados em uma string de consulta SQL, armazenada na varivel query. Aqui, o comando DELETE  usado para remover o registro de um cliente especfico, identificado pelo ID_Cliente=5008804.

 2. Estabelecendo conexo com o banco de dados: Usamos with engine.connect( ) as conn para abrir uma conexo com o banco. O with  muito til porque gerencia automaticamente a abertura e o fechamento da conexo. Isso significa que, aps o bloco de cdigo ser executado, a conexo ser fechada automaticamente, evitando que recursos sejam desperdiados ou bloqueios no banco de dados se prolonguem.

 3. Executando a consulta: Dentro do bloco with, executamos a consulta com conn.execute(text(query)). A funo text( )  usada aqui para garantir que a string de consulta seja interpretada como SQL puro.

 4. Confirmar a transao: Por fim, chamamos conn.commit() para confirmar a transao. Isso  crucial porque, em muitos bancos de dados, as alteraes feitas (como deletar um registro) no so salvas permanentemente at que sejam confirmadas explicitamente.

Este processo garante que o registro seja removido de forma segura e eficiente, minimizando a possibilidade de erros ou problemas de segurana.

# Atualizar o registro do cliente de ID 6840104 que teve o rendimento anual alterado para 300000

O cdigo a seguir atualiza o rendimento anual do cliente de ID 6840104 na tabela clientes. Para realizar essa tarefa, usamos uma instruo SQL UPDATE, que permite alterar os dados existentes em uma tabela.

A atualizao ser feita no campo Rendimento_anual, alterando o valor para 300000.0, mas isso s ser aplicado ao registro onde o ID_Cliente  igual a 6840104. O SQL para essa operao  construdo e armazenado na varivel query.

A execuo da query  feita utilizando uma conexo com o banco de dados criada pelo objeto engine. O cdigo segue estas etapas:

 1. Abrir a conexo: Usamos engine.connect( ) para estabelecer uma comunicao com o banco de dados.

 2. Executar a query: O mtodo conn.execute( ) executa a instruo SQL armazenada em query.

 3. Confirmar a alterao: O mtodo conn.commit( )  chamado para garantir que a modificao seja salva permanentemente no banco de dados. Sem isso, as alteraes podem no ser aplicadas.

```python
query = 'UPDATE Clientes SET Rendimento_anual=300000.0 WHERE ID_Cliente=6840104'
with engine.connect() as conn:
  result = conn.execute(text(query))
  conn.commit()
```

Quando trabalhamos com bancos de dados,  comum precisar inserir novos registros em nossas tabelas. Para isso, podemos utilizar a clusula INSERT INTO que nos permite adicionar novos dados de forma prtica e rpida.

Para utilizar essa clusula, primeiro  necessrio especificar o nome da tabela na qual o registro ser inserido. Em seguida, devemos passar, entre parnteses, os nomes das colunas onde sero adicionados os novos valores.  importante lembrar que a ordem dos nomes deve ser a mesma ordem em que os valores sero passados.

Aps especificar as colunas, utilizamos a clusula VALUES para adicionar os novos valores. Novamente, devemos passar entre parnteses os valores a serem inseridos na mesma ordem em que as colunas foram especificadas.  importante garantir que os valores estejam no formato correto para cada tipo de dado.

Como a query  longa voc pode usar ' \ para dizer que o cdigo continua na linha seguinte.

```python
query = 'INSERT INTO clientes (ID_Cliente, Idade, Grau_escolaridade, Estado_civil, ' \
        'Tamanho_familia, Categoria_de_renda, Ocupacao, Anos_empregado, ' \
        'Rendimento_anual, Tem_carro, Moradia) ' \
        'VALUES (6850985, 33, "Doutorado", "Solteiro", 1, "Empregado", "TI", ' \
        '2, 290000, 0, "Casa/apartamento prprio")'

with engine.connect() as conn:
  result = conn.execute(text(query))
  conn.commit()
```
