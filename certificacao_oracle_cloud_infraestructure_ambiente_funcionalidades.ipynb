## **Desafios do ambiente On-Premises**

Em um ambiente on-premises, que é um ambiente de infraestrutura local, enfrentamos desafios como infraestruturas caras e geralmente inflexíveis. Isso pode ocorrer devido a limitações de espaço, capacidade de processamento ou memória. Além disso, há cobranças expressivas e alta latência, dependendo da rede ou do provedor de nuvem utilizado. Muitas vezes, as faturas são imprevisíveis.

No OCI, conseguimos previsibilidade de valores, baixa latência e flexibilidade de recursos. Podemos aumentar ou diminuir recursos em produção, o que chamamos de ajuste "a quente". Embora reduzir recursos nem sempre seja fácil, aumentar geralmente é. Também temos load balance, entre outros recursos.

Segundo a documentação oficial, uma região é uma área geográfica que contém um ou mais datacenters, conhecidos como domínios de disponibilidade. Esses domínios são datacenters independentes dentro de uma mesma região, com recursos físicos isolados.

Dentro de um único domínio de disponibilidade, mesmo que seja apenas um, temos domínios de falha. Um domínio de falha é um agrupamento de hardware e infraestrutura dentro de um domínio de disponibilidade, completamente independente. Em um domínio de disponibilidade único, sempre teremos três domínios de falha.

A documentação menciona que teremos três domínios de falha para garantir a tolerância a falhas. Assim, entre vários domínios, podemos distribuir nossos recursos e máquinas dentro desse domínio de disponibilidade, divididos entre os domínios de falha. Caso um deles falhe, teremos nosso recurso ou outro recurso em outro domínio de falha.

Por exemplo, podemos ter nosso servidor principal dentro de um domínio de falha e um servidor de redundância, que é um secundário menor para poupar custos e recursos. Se o servidor principal falhar, o secundário, com menos recursos, assumirá, resultando em um serviço mais lento, mas que não cai.

   **Domínios de disponibilidade**

Os domínios de disponibilidade são data centers independentes e independentes dentro de uma região. Os recursos físicos de cada domínio de disponibilidade são isolados dos recursos de outros domínios de disponibilidade, o que fornece tolerância a falhas. Os domínios de disponibilidade não compartilham infraestrutura como energia ou resfriamento ou a rede interna de domínios de disponibilidade. Portanto, não é possível afetar os outros domínios de disponibilidade na região.

   **Domínios de falha**

Domínio de falha é um agrupamento de hardware e infraestrutura dentro de um domínio de disponibilidade. Cada domínio de disponibilidade tem três domínios de falha com energia e hardware independentes. Quando você distribui recursos entre vários domínios de falha, seus aplicativos podem tolerar falhas físicas do servidor, manutenção do sistema e falhas de alimentação dentro de um domínio de falha.

Os recursos dessa arquitetura são distribuídos igualmente entre os domínios de falha em cada domínio de disponibilidade.

Vamos entender como o IAM pode corrigir esses pontos.

## **Compreendendo os conceitos de autenticação e autorização**

Autenticação refere-se à confirmação da identidade da pessoa usuária no sistema, como o login em um e-mail ou no ambiente do curso, utilizando e-mail, senha e usuário. A autorização, por sua vez, define o que o usuário pode fazer no sistema. Por exemplo, um usuário comum tem certas permissões, enquanto um gerente ou administrador possui outras.

## **Enfrentando desafios e implementando soluções de acesso**

Dentro da OCI, a autenticação prova a identidade do usuário, e a autorização determina suas permissões. Os principais problemas relacionados a esses processos incluem acesso indevido a recursos sensíveis, uso de senhas fracas e roubo de senhas.

Para mitigar esses problemas, é importante definir padrões de senha, como tamanho e formatação, e exigir a troca periódica de senhas. A OCI possui uma política que obriga a troca de senha após um determinado período.

Além disso, a dificuldade de gerenciar acessos entre sistemas diferentes, como infraestrutura local (on-premise), e a necessidade de auditoria e controle de acesso são desafios a serem enfrentados. Para garantir que o usuário é quem diz ser, utilizamos login e senha, além do multifator de autenticação (MFA), que pode incluir o uso de aplicativos como o Google Authenticator.

A autorização é garantida por meio de políticas do IAM, que definem as ações permitidas para cada usuário. A autenticação (AuthN) verifica a identidade do usuário, enquanto a autorização (AuthZ) define suas permissões. O IAM gerencia ambos os processos. No contexto da OCI, trabalhamos com políticas aplicadas a usuários, grupos e compartimentos, permitindo a segregação de acessos.

# **Tenancy**

Estamos visualizando o console da OCI aberto novamente. Agora, discutiremos sobre Tenancy.

## **Compreendendo o que é a Tenancy**

Para demonstrar o que é Tenancy, vamos abri-lo. Quando fazemos login na conta da OCI, é necessário fornecer o identificador da Tenancy. Isso nada mais é do que a conta principal. Ao criar uma conta na OCI, ela se torna a Tenancy. Ela recebe um OCD, por exemplo, e contém todas as informações da Tenancy.

A Tenancy é a conta mestre dentro da organização da OCI, sendo a unidade principal de gerenciamento da Oracle, onde todos os recursos são organizados. Ela é a estrutura principal. Quando trabalhamos com os compartments, observamos que todos eles são filhos de um primeiro, que é, consequentemente, a nossa Tenancy.

A Tenancy abrange tudo: identidade e acesso (nosso IAM), recursos em nuvem, faturamento e monitoramento, governança e segurança. A Tenancy é o que engloba todo o restante, sendo realmente o principal.

## **Identificando problemas de configuração e organização**

Entre os problemas decorrentes da não utilização adequada ou da configuração incorreta da Tenancy, podemos temos a falta de organização, controle de recursos e acesso sem restrição, questões já abordadas em vídeos anteriores, além da dificuldade na gestão de múltiplas equipes e projetos. ## **Aplicando soluções e boas práticas na Tenancy**

A solução dentro desse contexto, segundo a explicação sobre Tenancy, é bastante rápida, pois se trata da conta principal. No entanto, o ponto principal são as boas práticas.

Primeiro, temos o Tenancy Admin, que é a conta principal ou conta mestre. Em seguida, criamos um administrador dentro da OCI, chamado OCI Admin. Portanto, criamos um OCI Admin e, mesmo sendo um administrador, restringimos algumas funcionalidades. Isso evita problemas, inclusive em alterações na própria Tenancy.

## **Estruturando grupos, políticas e compartments**

Após criar o OCI Admin, formamos o OCI-admin-group. Nesse grupo, podemos definir políticas conforme desejarmos. Existem administradores de banco de dados, instâncias e VCNs (Virtual Cloud Networks). Podemos dividir e segregar os recursos conforme necessário.

Por último, mas não menos importante, temos a implementação da política. Consequentemente, criamos o Compartment para isolar recursos. Ao isolar recursos, isolamos responsabilidades e tipos de problemas, Por exemplo, definimos quem poderá trabalhar com o Compartment Test 1 e, dentro do Compartment 1, quem poderá trabalhar com instâncias, bancos de dados e afins. Essa estrutura permite definir essas responsabilidades.

Além disso, implementamos a Autenticação Multifator (MFA) para reforçar a segurança na autenticação. A configuração da Tenancy é a estrutura principal, onde devemos seguir rigorosamente para garantir auditoria, conformidade e outros aspectos importantes.

# **Aula 03 - VRede de Nuvem Virtual (VCN)**

# **Apresentando o conceito de OCI Virtual Cloud Network (VCN)**

Temos a OCI Virtual Cloud Network, conhecida como VCN, um serviço fundamental dentro da OCI que permite criar e gerenciar redes privadas e públicas de forma isolada, considerando todo o IAM, compartment, etc., que são aspectos interessantes. Este é um passo seguinte, por exemplo, após a criação do compartment. Depois de criar os compartimentos e dividi-los, criamos as redes de cada um.

A VCN é uma rede definida por software, não sendo algo físico que se divide em servidores separados. Ela é dividida pelo software, sendo totalmente personalizável, com baixa latência, alta disponibilidade e totalmente escalável.

Atualmente, sem o uso dessas redes distribuídas, enfrentamos acesso não autorizado e falhas de segurança. Com a VCN, podemos definir o que entra em cada local, até onde pode ir, entre a rede pública e a privada. Além disso, enfrentamos baixo desempenho e alta latência, especialmente em ambientes locais, conhecidos como on-premises. Há também dificuldade em conectar ambientes híbridos, multicloud, e, consequentemente, em escalar. A escalabilidade em um ambiente físico dentro de uma organização é muito mais complexa.

## **Criando uma VCN manualmente**

No canto superior esquerdo, no menu sanduíche, acessamos "Networking" > "Virtual cloud networks" para acessar a VCN. Vamos fazer uma apresentação e, em seguida, criar uma VCN juntos.

Primeiro, acessarei minha VCN, chamada vcn-espirito-santo. Dentro desta VCN, temos as informações sobre o compartment em que foi criada e a data de criação. Chegamos, então, ao ponto importante, a notação CDR, que é o espaço de endereço. Isso é básico de rede, e não vamos nos aprofundar, pois quem já tem um estudo mais avançado em redes sabe que existem calculadoras de subrede na internet, eliminando a necessidade de cálculos manuais. Não falaremos sobre como se chega a números como /16 ou /24, mas é importante entender como a VCN funciona.

Na VCN, temos um bloco específico definido em 10.0.0.0/16, e duas subnets: a privada e a pública. A pública mantém o mesmo espaço 10.0.0.0, mas com /24, enquanto a privada tem um 1 para separar: 10.0.1.0/24.

Se todas as máquinas estiverem configuradas como privadas, dentro de uma rede privada, é necessário implementar controles para o acesso a recursos externos. Uma das soluções possíveis é direcionar o tráfego por um load balancer, que abordaremos na próxima aula. Esse mecanismo permite que o tráfego atinja primeiro a rede pública e, em seguida, a rede privada.

## **Entendendo os tipos de gateway e roteamento**

Ao tratar de VCN, é essencial entender como ocorre sua conexão com o ambiente externo. Um dos recursos disponíveis é o Internet Gateways, que permite conectividade direta com a internet. Já o NAT Gateways oferece o serviço de NAT de forma unidirecional, possibilitando apenas o tráfego de saída — ou seja, as subnets podem acessar recursos externos, mas não podem ser acessadas de fora.

Por fim, há o Service Gateways, que é utilizado quando precisamos que recursos dentro da VCN acessem outros serviços públicos da OCI. Por isso, utilizamos o Gateway como um serviço. Além disso, temos o Gateway de Roteamento Dinâmico (Dynamic Routing Gateway Attachments), que é um roteador de tráfego privado da nossa VCN. Supondo que desejamos direcionar o tráfego para outros destinos na internet, especificamos como queremos trabalhar com esse roteador dinâmico.

## **Criando uma VCN com o assistente (Wizard)**

Agora que explicamos os principais pontos da subnet, incluindo os serviços de Gateway, roteamento e o Internet Gateway, vamos voltar à seção de Virtual Cloud Networks. Não vamos clicar em "Create", mas sim trabalhar com o Wizard da OCI.

Ao escolher a opção "Start VCN Wizard", temos a opção de criar uma VCN para conectar à internet ou para VPN. No entanto, não queremos uma VPN. Se desejarmos conectar, podemos ver a opção "On-Premise Network". Isso é útil para conectar uma estrutura local com uma estrutura na nuvem, especialmente em ambientes multicloud.

Hoje, vamos falar sobre o Wizard da criação padrão. Apesar de estarmos com o limite da conta gratuita já atingido, isso não será um problema. Para criar uma VCN, inserimos, por exemplo, "teste-001" e escolhemos o compartment "curso-oci".

Após criar o compartment, podemos editar o CDR, o bloco de IPs, conforme necessário. Podemos optar por trabalhar com IPs da faixa 10, 192 ou 172, que são reservados para ambientes privados. As alterações podem ser feitas conforme necessário.

Se o bloco de IP tiver muitos IPs disponíveis, podemos ajustar o cálculo da subrede conforme desejado. Vamos deixar tudo no padrão e clicar em "Next". O sistema nos informará sobre o upgrade e apresentará um resumo de tudo que será criado.

Ao final, basta clicar em "Create" para que o sistema crie tudo automaticamente. Após a criação, teremos nossa VCN pronta, com todas as conectividades importantes configuradas. O NAT Gateway não é criado, pois não é necessário para acessar a internet. Ao criar uma instância, escolhemos se ela terá um IP público ou não, mas isso será abordado na seção de instâncias. ## **Entendendo a funcionalidade e importância do load balancer**

O Load Balancer é um serviço da Oracle que distribui automaticamente o tráfego de múltiplos servidores. Ele funciona como um serviço de alta disponibilidade, otimizando a performance e oferecendo segurança avançada.

A otimização de performance ocorre porque ele distribui eficientemente o tráfego, evitando problemas como servidores sobrecarregados e falta de redundância. Em situações como um e-commerce durante promoções de final de ano, a ausência de um balanceador de carga pode resultar em servidores paralisados e distribuição ineficiente de tráfego.

Para resolver esses problemas, utilizamos um balanceador de carga. No painel, ao clicar em "Load Balancer", temos dois tipos principais: o Load Balancer flexível e o dinâmico. O flexível permite parametrizar um ponto mínimo e máximo de operação, enquanto o dinâmico já possui pré-configurações para realizar o balanceamento automaticamente.

Podemos escolher entre um Load Balancer público ou privado. O público é utilizado em casos de servidores que recebem tráfego diretamente da internet, como em promoções de final de ano. Já o privado é ideal para balancear a carga de bancos de dados que não precisam estar disponíveis na internet.

O segundo tipo de Load Balancer é o Network Load Balancer, que opera principalmente nas camadas 3 e 4 do modelo. Ele é mais rápido e é utilizado quando a velocidade e disponibilidade são prioridades. Se precisarmos de mais inteligência no balanceamento de carga, optamos pelo Load Balancer tradicional.

# **Aula 04 - Introduzindo o Compute**

Vamos para mais uma etapa. Agora, discutiremos sobre computação e, consequentemente, as instâncias.

O que são compute e instances dentro da OCI? Vamos abordar máquinas virtuais (Virtual Machines). Podemos nos referir a elas como VMs devido ao termo Virtual Machine. Também falaremos sobre servidores Bare Metal. Esses são os dois tipos que abordaremos mais adiante.

## **Listando vantagens da computação**

Quando entramos no tema da computação, podemos falar sobre flexibilidade. A flexibilidade permite escolher como trabalhar, seja com máquinas virtuais ou servidores Bare Metal, dependendo do desempenho necessário.

A escalabilidade é outro diferencial, pois a criação ou remoção de instâncias é fácil e rápida. Em termos de segurança, temos controle total sobre a parte de Firewall, VCN, compartments, tudo que já discutimos em outros vídeos.

Além disso, possui boa customização. Permite escolher diferentes sistemas operacionais, quantidade de CPU, memória e armazenamento.

## **Resolvendo problemas com computação na OCI**

Quais problemas existentes que podemos resolver com a computação na OCI?

 - Alto custo de hardware e manutenção;

 - Dificuldade para escalar rapidamente;

 - Baixo desempenho e limitações técnicas;

 - Complexidade no gerenciamento.

O alto custo de hardware e manutenção é um deles. Manter um servidor físico tem um custo elevado de manutenção e segurança. É necessário manter toda uma estrutura, não apenas o servidor, mas também sala-cofre, chaves, acesso, etc.

A dificuldade para escalar rapidamente é outro problema. Criar ou reduzir uma máquina nova é complicado. O baixo desempenho e limitações técnicas também são desafios. No ponto de inteligência artificial, por exemplo, não há algo próprio. Na OCI, já existem aplicações específicas que podem ser facilmente integradas às máquinas.

Outro ponto é a complexidade no gerenciamento. Administrar servidores físicos requer mais tempo, enquanto trabalhar apenas com instâncias prontas envolve apenas a parte de software.

## **Exemplificando uso real**

Vamos trazer um exemplo real. Uma fintech precisava aumentar rapidamente sua capacidade de processamento durante picos de transações financeiras, mas seus servidores locais não conseguiam escalar automaticamente, o que resultou em lentidão e falhas no sistema.

Este é um exemplo de load balance, que na computação em nuvem da OCI é fácil de implementar, como exemplificado em vídeos anteriores. No ambiente on-premises, ou infraestrutura local, isso é complicado de realizar.

A solução tradicional envolve a compra de novos servidores, mas isso pode levar semanas para chegar e tem um custo alto, além de demandar vários profissionais para estruturar a rede.

A solução ideal é utilizar a OCI Compute, a parte de computação, para criar instâncias sob demanda, pagando apenas pelo uso. Em um servidor local, isso não é possível. Se compramos um disco de 2TB, pagamos por 2TB. Na OCI, se usamos 150GB, pagamos por 150GB.

Dentro da solução ideal, contamos com a criação rápida de servidores, alto desempenho, suporte para vários tipos de processadores, como AMD e Intel. A escalabilidade dinâmica permite aumentar ou reduzir recursos, seja de processador ou memória, de forma fácil.

A segurança e o isolamento total são garantidos pelas VCNs e subnets privadas e públicas, proporcionando um ambiente mais cômodo. Além disso, há o isolamento dos compartments.

## **Criando instância na OCI**

Vamos acessar nosso ambiente da OCI, o console. No canto superior esquerdo, acessamos o menu sanduíche, que é o ícone de três traços. Vamos navegar até a opção "Compute > Instances". Assim, poderemos conferir as instâncias existentes, usadas anteriormente para a criação do Load Balancer.

Vamos clicar no botão "Create Instance". Dentro dessa opção, podemos escolher o nome da instância e o compartment onde será criada. Nesse exemplo, manteremos todos os valores sugeridos.

Também podemos escolher o domínio de disponibilidade. Para isso, na seção "Placement", vamos clicar em "Edit" e "Show advanced options". Para esse curso em específico, manteremos a opção "on-demand capacity", ou seja, por demanda de capacidade.

No dropdown "Fault domain", podemos mudar o domínio de falha. Mas, manteremos a opção "Let Oracle choose best fault domain" para que a Oracle escolha o melhor domínio.

Além disso, não alteramos nada na parte de segurança.

Na seção "Image and Shape", o Oracle Linux é a opção padrão de imagem, mas podemos escolher entre Oracle Linux, Ubuntu, Red Hat, CentOS, Windows, entre outros, incluindo imagens customizadas. Por exemplo, podemos mudar para Ubuntu e escolher entre suas versões.

O shape refere-se à estrutura escolhida. Temos a opção de máquina virtual e bare metal, como mencionado anteriormente. A máquina virtual é uma máquina comum, onde podemos escolher entre processadores AMD, Intel e Ampere (que são processadores móveis de alta potência).

O bare metal é uma máquina física dedicada. Ao invés de utilizar uma máquina virtual apenas por meio de um software, teremos uma máquina inteira dentro do datacenter, exclusivamente nossa.

Quando falamos sobre instâncias, é importante mencionar que, dentro de cada instância que configuramos, já são criados um compartment e uma VCN. Estamos dentro de um domínio de disponibilidade, dentro de um compartment, dentro de uma VCN, e, consequentemente, dentro da tenancy e afins.

Na configuração de VCN, além da opção de criar uma nova ou selecionar uma existente, existe a opção de VNIC. VNIC permite dar um nome específico à placa de rede. Assim, podemos nomear a placa de rede física do computador para algumas configurações futuras, mas não vamos nos aprofundar nisso agora.

Na parte de endereço IP primário da VNIC, podemos fazer a atribuição automática de IPv4 privado e público. Se a rede for pública, um IP público será criado automaticamente. Caso seja privada, o IP público não será criado. Nós temos um IP público para fazer uma conexão de exemplo mais rápida.

Essas máquinas terão, em sua estrutura, um volume anexado a elas. Na OCI, é comum ouvir os termos attach e detach, que significam fixar ou desfixar um disco. Vamos explorar isso mais adiante. Portanto, o boot volume é o volume de inicialização, já o block volume funciona como um HD ou SSD secundário.

Na parte de "SSH Keys", podemos criar um novo par de chaves e fazer o download para acessar via SSH posteriormente. # **Scaling e OKE**

Vamos abordar dois assuntos: Scaling (escalonamento) e OKE.

## **Entendendo o que é Scaling**

O Scaling é a capacidade de aumentar ou reduzir recursos computacionais conforme a demanda da aplicação, de forma manual ou automática.

O primeiro tipo é o Vertical Scaling, onde conseguimos aumentar ou reduzir recursos como CPU e memória. O Horizontal Scaling, também conhecido como Auto Scaling, permite adicionar ou remover instâncias de forma automática. Vamos explorar mais adiante como isso funciona exatamente.

    No que se refere ao redimensionamento vertical, uma boa prática é que ele seja planejado - para que seja realizado fora do horário de produção.

Mesmo que seja possível fazer em produção a qualquer momento, isso pode gerar um downtime, deixando o servidor offline por alguns segundos ou minutos. Isso pode ocasionar perdas de dados, pois interromperia operações ativas. Portanto, é uma boa prática realizar o redimensionamento de forma planejada, mesmo que seja possível fazê-lo com o servidor ligado.

## **Resolvendo problemas com escalonamento**

Vamos discutir alguns problemas que podem ocorrer com a não utilização do Scaling.

 - Recursos insuficientes em picos de demanda;

 - Desperfício de recursos;

 - Dificuldade em gerenciar aplicações distribuídas;

 - Tempo de resposta alto.

Suponhamos que estejamos trabalhando em um ambiente de infraestrutura local (on-premise) e não tenhamos recursos suficientes em um pico de demanda. Em outro cenário, podemos ter muitos recursos e pouco uso, resultando em desperdício de recursos.

Além disso, podemos enfrentar dificuldades em gerenciar aplicações distribuídas, como microsserviços. Por fim, o tempo de resposta tende a ser mais alto, especialmente em termos de latência em servidores, em datacenters gerenciados localmente.

## **Exemplificando uso real**

Agora, vamos apresentar um exemplo real. Imagine que estamos operando uma loja virtual e, de repente, o tráfego aumenta ou ocorre a indisponibilidade de alguma máquina. Infelizmente, estamos sujeitos a isso.

Nesse momento, conseguimos de forma automática, com o scaling horizontal, adicionar ou reduzir instâncias conforme a necessidade. Se o tráfego aumentar, aumentamos a instância; se o tráfego diminuir, reduzimos a instância.

Como fazemos isso? Inicialmente, precisamos de uma instância rodando. Com uma instância, o primeiro passo é criar o template, isto é, uma imagem completa da instância, semelhante a um snapshot em alguns sistemas, que é como tirar uma foto.

Após isso, criamos o que chamamos de pool de instâncias, que é uma coleção de instâncias. Com a coleção de instâncias a partir dos templates, escrevemos as regras de scaling horizontal.

Por exemplo, se a memória for maior que 70%, queremos duas máquinas, ou seja, ele sobe uma segunda instância de forma automática. Se for menor que 20%, queremos apenas uma, então fazemos o downgrade para reduzir o recurso.

## **Explorando OK**

No segundo momento, vamos discutir o OKE (Oracle Kubernetes Engine), que é um orquestrador, um serviço gerenciado de Kubernetes dentro da OCI, com alta disponibilidade, alto desempenho, eficiência, segurança e isolamento.

Vamos agora discutir a diferença entre trabalhar com uma máquina virtual (VM) e um container. Quando falamos de máquina virtual, temos o hardware, um software virtualizador e, consequentemente, as máquinas virtuais - cada uma com seu próprio sistema operacional, aplicações e afins.

O container, por outro lado, possui o hardware e o sistema operacional, e dentro dele, um Docker, que é o serviço que trabalha com containers. Com o Docker instalado no sistema operacional, ele gerencia os containers. O container precisa de apenas algumas bibliotecas para ser executado.

Por que usar um ou outro? Quando falamos de máquina virtual, ela possui CPU, memória, disco e outros recursos separados. O container é iniciado muito mais rapidamente, pois não precisa carregar uma gama completa de recursos. Ele é portátil e eficiente.

Qualquer lugar em que tivermos nosso script de criação do container, se soubermos qual vamos criar, conseguimos levá-lo para qualquer local.

Na prática, com os containers rodando, precisamos de algo a mais para realizar toda a gerência deles. Para orquestrar esses containers, podemos usar os Kubernetes.

    O Kubernetes é um sistema de código aberto (open source) e é utilizado para gerenciar os containers.

Quando falamos em gerenciar, referimo-nos à implantação, dimensionamento e gerenciamento completo. As vantagens incluem o controle do escalonamento e a simplificação das operações, o que proporciona mais resiliência e elimina o tempo de inatividade, conhecido como downtime.

Em resumo, o Docker faz a gerência do container, enquanto o Kubernetes ajuda na orquestração e coordenação. Esses dois trabalham juntos.

# **Listando tipos de clusters no OKE**

O OKE é totalmente gerenciável, escalável e altamente disponível. Entendido isso, podemos conhecer os dois tipos de cluster dentro do OKE.

O primeiro tipo é conhecido como cluster aprimorado, que oferece todos os recursos e um SLA (service level agreement). O SLA é um acordo entre um fornecedor e um cliente pagante. Isso será financeiramente respaldado e, consequentemente, terá a garantia de que todos os serviços contratados serão entregues.

O segundo tipo é conhecido como cluster básico, ou seja, possui funções básicas sem os recursos aprimorados. O básico tem um SLO (service level objective), em vez de um SLA. O SLO é um objetivo de nível de serviço. O SLO são promessas, sem pagamento, portanto, sem garantia. Apenas promessas de que os serviços e afins ocorrerão.

Por fim, vamos entender o que são nós de cluster. No nó de cluster, temos o nó virtual e o nó gerenciado. O nó virtual é a opção sem servidor, gerenciado completamente pelo CI, e só pode ser utilizado dentro de um cluster aprimorado. Já o gerenciado, você terá que gerenciar esses nós.

Assim, se você escolher o aprimorado, você consegue usar o nó virtual que é gerenciado pelo CI. Se você optar pelo básico, você irá usar o nó gerenciado e toda a gerência será feita por você.

# **Aula 05 - Introduzindo o Storage**

Vamos começar a discutir sobre Storage (armazenamento) na OCI.

O Storage é um armazenamento escalável e seguro para nossas aplicações. Ou seja, é um serviço onde armazenamos, gerenciamos e acessamos dados de forma segura, rápida e altamente escalável dentro do ambiente da OCI.

##**Entendendo desafios do armazenamento local**

Os problemas que enfrentamos ao não utilizar esse tipo de infraestrutura incluem o alto custo com armazenamento local que chamamos de on-premise. Manter nosso próprio datacenter ou servidor resulta em custos elevados, especialmente no investimento necessário para manter uma estrutura de backup e lidar com grandes volumes de dados.

Imaginemos um cenário com 10 terabytes de dados; para replicar, seriam necessários 20 terabytes, considerando o backup. Se fizermos mais uma cópia, chegamos a 30 terabytes, o que é complicado para armazenamento on-premise em uma infraestrutura local.

Além disso, o acesso a dados arquivados há muito tempo pode ser lento. Muitas vezes, temos sistemas onde certos dados não são acessados com frequência, e quando precisamos deles, o acesso é demorado.

Analisando critérios essenciais do armazenamento
Para explorar o assunto de storage, precisamos considerar alguns pontos importantes:

 - O tipo de armazenamento;

 - A performance;

 - A capacidade;

Em relação ao tipo de armazenamento, devemos definir se será persistente ou não persistente. Para a performance, decidiremos qual nível de performance será necessário. Para a capacidade, decidiremos o quão performática será a escolha de armazenamento.

Outro aspecto importante é a durabilidade, na qual nos referiremos às cópias de dados que serão mantidas.

Talvez você já tenha ouvido o termo retenção. A diferença entre persistência e durabilidade é que a primeira refere-se à segurança no armazenamento dos dados, enquanto a segunda diz respeito ao tempo de armazenamento dessas cópias.

##**Decidindo sobre conectividade e localização**

Além disso, devemos considerar a conectividade, decidindo onde armazenar os dados:

 - Em um *local storage* (armazenamento local) ou

 - Em um *network storage* (armazenamento em rede).

Também consideraremos como nossa aplicação acessará esses dados e qual protocolo será utilizado. Esses são pontos importantes a serem considerados para escolher o serviço ideal para nossa aplicação ou serviço.

##**Explorando tipos de armazenamento**

Atualmente, temos os seguintes tipos de armazenamento:

 - Local NVMe;

 - Block Volume;

 - File Storage;

 - Object Storage

O armazenamento Local NVMe possui uma estrutura dentro de um domínio de disponibilidade, com um servidor e armazenamento conectado diretamente a ele, utilizando SSDs de alta velocidade. Esse tipo de armazenamento é indicado para aplicativos que exigem alto desempenho.

Temos também o **Block Volume**, ou Volume de Bloco. Ele começa de forma semelhante ao armazenamento local, mas o armazenamento é **conectado via rede**. Isso o torna persistente e durável, pois ele dura além da instância, já que não está diretamente na instância, mas conectado a ela via rede.

Quando mencionamos instância, estamos nos referindo, por exemplo, a uma máquina virtual. Assim, podemos ter a máquina virtual conectada ao volume de bloco.

O **File Storage** (armazenamento de arquivos) é um armazenamento compartilhado. Suponhamos que temos duas instâncias (como duas máquinas virtuais) e ambas compartilham os mesmos dados. O File Storage opera na estrutura de pastas e arquivos. Se precisarmos trabalhar dessa forma, o File Storage é uma boa escolha.

O **Object Storage** (armazenamento de objetos) é voltado para a web. Ele é utilizado, por exemplo, para armazenar fotos, vídeos e áudios, que são acessados por um cliente via internet, por exemplo. Em outros vídeos, já discutimos o uso do Object Storage para tratamento de dados.

##**Solucionando a migração de dados**

Além disso, temos ambientes de migração dentro dos módulos de Storage. O **Data Transfer Disk** é onde enviamos nossos discos para que a OCI realize a migração de dados. O **Data Transfer Appliance** utiliza um Appliance maior para enviar dados para a OCI.

Por fim, o Storage Gateway é um servidor Linux utilizado para migrar dados para a OCI, auxiliando nesse processo de migração.

##**Trabalhando com objetos**

O *Object Storage* é um armazenamento de alto desempenho e escala da internet. Ele trabalha diretamente com objetos, e, consequentemente, todos os dados serão armazenados e gerenciados dessa forma.

Independentemente do tipo de dado, ele será tratado como objeto. É ideal para dados principalmente não estruturados, como logs, vídeos, arquivos e imagens.

Além disso, todo o valor do conteúdo e cada objeto receberá seus metadados.

    Metadados são dados sobre dados.

##**Armazenando e acessando os objetos**

Os buckets ("baldes", em português) armazenam os objetos. Eles possuem um nome único dentro da nossa tenancy e estão em uma hierarquia plana, divididos apenas por prefixos.

Temos também o conceito de namespace, uma entidade lógica atuando como um contêiner de nível superior para todos os buckets da nossa conta. Ele possui um nome global único.

Para acessar esses objetos, utilizamos um endpoint, o ponto final de uma URL que direciona a uma parte do site. Por exemplo, uma URL do site oficial da OCI pode ser algo como:

  https://swiftobjectstorage.us-ashburn-1.oraclecloud.com/n/exemplo/n/exemplo2/arquivo.zip

Nesse caso, o arquivo é um .zp, usado apenas como exemplo, sem representar um tipo de arquivo específico.

O início da URL do serviço, localizado na região leste dos Estados Unidos, é https://swiftobjectstorage.us-ashburn-1.oraclecloud.com. O restante do caminho vai sendo completado com as informações do objeto armazenado.

Esse objeto é identificado pelo sufixo /n/, e no final, o /arquivo.zip. Além disso, existe o namespace exemplo, que também compõe a URL final da API do Object Storage.

##**Explorando níveis de armazenamento**

Em relação ao armazenamento, temos os seguintes níveis:

 - Standard;

 - Infrequent Access Tier;

 - Archive Tier;

 - Auto-Tiering.

O *Standard* é utilizado para **dados críticos** que precisam ser acessados frequentemente. O *Infrequent Access Tier* é para **dados** importantes que **não** precisam ser acessados frequentemente.

Há também o *Archive Tier* (tipo de arquivo), para dados que **raramente** são acessados. Por fim, o *Auto-Tiering* organiza de acordo com o nível de acesso necessário, variando entre o Standard e o Infrequent, conforme o uso.

Quando criarmos nosso bucket, veremos esses níveis em ação.

##**Entendendo e classificando os Block Volumes**

O Block Volume, em sua essência, é um armazenamento persistente e durável para instâncias de computação, hospedado em um servidor de rede. Por exemplo, é possível criar e anexar volumes sem depender da vida útil da instância, pois eles são independentes dela.

O Block Volume funciona destinado a instâncias de computação, oferecendo armazenamento em bloco. Podemos criar e anexar esses discos às instâncias, assim como desanexá-los e excluí-los.

É importante ressaltar que esses dados são completamente independentes. Se um Block Volume estiver "attachado" (anexado) a uma instância e a instância for apagada, o Block Volume permanece independente.

Eles são persistentes, o que significa que permanecem armazenados de forma segura, mesmo após a exclusão da instância. Além disso, são duráveis, pois o serviço mantém múltiplas cópias dos dados.

Vamos agora discutir os tipos de Block Volume, que são:

 - Baixo custo;

 - Equilibrado;

 - Alto desempenho;

 - Ultra Alto desempenho.

O tipo de baixo custo é geralmente utilizado para streaming de dados, com processamento mais lento. O tipo equilibrado é usado, por exemplo, para discos de inicialização de sistemas.

O de alto desempenho é projetado para alta demanda, e o de ultra alto desempenho é ideal para bancos de dados relacionais.

Um recurso avançado que veremos na demonstração é o Auto-Tune Performance, que ajusta automaticamente o desempenho do volume. Se um volume for desanexado, ele passa para um nível menor de custo. Quando é reanexado, ele retorna ao nível anterior, ajudando a reduzir custos.

##**Garantindo segurança e ampliando recursos**

Vamos falar sobre segurança e criptografia. Ao criar um Block Volume, a **criptografia não pode ser desativada**. É possível inserir **chaves próprias** no momento da criação.

Além disso, os dados são criptografados em repouso e durante a transmissão, quando há comunicação entre o serviço e a instância.

O compartilhamento de discos permite que um único Block Volume seja compartilhado com várias máquinas virtuais, desde que isso esteja bem estruturado na aplicação. O redimensionamento permite expandir os volumes de forma online, sem desanexar da máquina virtual, ou de forma offline, necessitando desanexar para a expansão.

A replicação do Block Volume oferece replicação assíncrona para outras regiões, como São Paulo e Estados Unidos, facilitando a expansão de negócios, migração de dados entre regiões e recuperação de desastres. Há também o **Volume Groups**, que permite agrupar múltiplos volumes dentro de um único Block Volume, facilitando a criação de backups.

O File Storage é, de forma simples, uma coleção hierárquica de documentos organizados em diretórios nomeados. Temos familiaridade com esse conceito, pois, no nosso dia a dia, organizamos arquivos em pastas e diretórios, tanto no Linux quanto no Windows.

# **Introduzindo o Database**

##**Introduzindo banco de dados na OCI**

Banco de dados é um conjunto de serviços gerenciados que permite armazenar, processar e analisar grandes volumes de dados de forma segura, escalável e altamente disponível.

Vamos explorar os principais serviços de banco de dados presentes na OCI.

 - **Oracle Autonomous Database**: um banco de dados totalmente gerenciado que se otimiza e se protege automaticamente;

 - **Oracle Exadata Cloud Service**: uma plataforma de alto desempenho para bancos de dados corporativos;

 - **MySQL HeatWave**: o MySQL já existe dentro da OCI como um banco específico, enquanto o HeatWave é uma engine otimizada para análises muito mais rápidas dentro da OCI.

##**Resolvendo problemas de bancos de dados on-premises**

Quais são os problemas que existem no mundo dos bancos de dados quando não utilizamos esses bancos na nuvem dentro na OCI? Quando usamos infraestrutura local, ou on-premises, enfrentamos dificuldades na administração e otimização. Isso exige configuração manual, além de otimização, backup e análise constante de patch de segurança.

Além disso, enfrentamos **baixo desempenho, alta latência e escalabilidade limitada**. Isso ocorre quando há muito recurso e pouca utilização, ou muita utilização e pouco recurso. Durante picos, não conseguimos escalar de forma fácil e dinâmica, como, por exemplo, nas promoções de final de ano de um site de e-commerce.

Além disso, há a **falta de automação e segurança**. Às vezes, não contamos com automação nativa e tudo é feito de forma manual - o que abre espaço para erros humanos.

##**Exemplificando uso real**

Um exemplo real disso é o processamento inesperado de milhões de transações por segundo, para o qual nossa plataforma ou hardware não estava preparado.

A solução tradicional seria adquirir mais servidores. Mas caso seja uma necessidade urgente, pode não ser possível encomendar e ampliar a capacidade tão rapidamente. Se você já tiver um hardware, ainda assim será necessário realizar ajustes manuais e alguém precisará dedicar tempo a isso.

A solução ideal é a migração para o Oracle Autonomous DB, que automatiza todo o processo, como backup e otimização de segurança, sem necessidade de intervenção humana.

Além disso, possui escalabilidade sob demanda, ou seja, permite aumentar e diminuir recursos conforme a necessidade. Também oferece segurança avançada, proteção contra ataques, criptografia e auditoria contínua. E, consequentemente, uma vantagem é a baixa latência e alto desempenho.

# **Autonomous DB**

Ddando continuidade ao assunto de banco de dados, vamos falar sobre o Autonomous Database, um banco de dados totalmente gerenciado, autônomo e escalável. Ele é projetado para eliminar tarefas administrativas complexas e, muitas vezes, repetitivas. Utilizando inteligência artificial e automação, ele otimiza o desempenho, segurança e disponibilidade - sem a necessidade de intervenção humana.

##**Caracterizando Autonomous Database**

As principais características incluem ser autogerenciado. Em outras pelavras, realiza o que chamamos de tuning automático, ou seja, se autoconfigura e gerencia seus próprios recursos.

Além disso, também é autoprotegido. Isto é, realiza atualizações, executa patches de segurança e faz backups. Também é autorreparável, pois localiza e repara falhas sem impacto para o usuário.

Finalmente, oferece autodesempenho, escalabilidade e baixo custo operacional. A ausência de necessidade de intervenção humana ajuda a reduzir o custo operacional, já que o papel do DBA (administrador do banco de dados) não é mais obrigatório.

##**Resolvendo problemas com Autonomous Database**

Entre os problemas que o Autonomous Database resolve estão a administração complexa e cara, risco de falha em disponibilidade e falha de segurança. Um problema específico que ele resolve é a falta de escalabilidade dinâmica. Afinal, o Autonomous Database faz tudo de forma automático e dinâmica.

As soluções que ele oferece incluem gerenciamento totalmente automático, escalabilidade dinâmica, proteção contra ameaças, alta disponibilidade e recuperação rápida.

##**Configurando Autonomous Database no console da OCI**

Agora, vamos para o console da OCI. Vamos expandir o menu de navegação lateral painel e clicar em "Oracle Database > Autonomous Database". Desse modo, acessaremos o Autonomous Database que criamos anteriormente, e que vamos usar agora.

Vamos clicar no botão "Create Autonomous Database" para começar a criação. No campo "display name", podemos nomeá-lo como teste001. Esse será o nome que aparecerá. No campo "database name", colocaremos db001, por exemplo. Vamos criá-lo no compartment curso-oci.

Em seguida, podemos escolher o tipo de workload, que define o tipo do nosso banco de dados. Por padrão, o Data Warehouse já vem selecionado. Esse tipo serve para autoprocessamento e operações de Big Data. O Transaction Processing é utilizado em sistemas OLTP e aplicações empresariais. Vale ressaltar que, em uma estrutura dedicada a Exadata, o JSON e o APEX não são suportados.

Na parte de configuração do banco de dados (Database configuration), a opção Always Free já está marcada. Ao desmarcá-la, é possível escolher a quantidade de ECPUs e, consequentemente, a quantidade de armazenamento (storage) desejada.

Conforme discutido anteriormente, temos as opções de Computação em Auto Scaling e Storage em Auto Scaling, o que permite que o sistema gerencie a escalabilidade de forma automática. Vamos manter a opção Always Free marcada, ou seja, sempre gratuito.

Na seção de criação de credenciais de administrador (Administrator credentials creation), é necessário inserir uma senha que atenda aos requisitos de segurança do Autonomous DB. Deve ter entre 12 a 30 caracteres e incluir pelo menos uma letra maiúscula, uma minúscula e um número.

No acesso à rede (Network Access), podemos definir o tipo de acesso que será liberado. Podemos permitir qualquer acesso da internet ou especificar alguns endereços de IP. Vamos deixar "Secure acess from everywhere" a título de exemplo.

Na seção de contato de notificação de manutenção, é possível inserir um e-mail de contato. É importante utilizar um e-mail que será monitorado para receber informações sobre manutenções programadas.

Em seguida, vamos clicar em "Show advances options" no final da página para exibir as opções avançadas. Na aba "Encryption key", encontramos as chaves. Podemos optar por deixar a gerência das chaves a cargo da própria UCI ou trazer nossas próprias chaves, que devem estar dentro de um cofre (vault). O vault será abordado em um vídeo futuro.

Após configurar tudo, podemos confirmar nossas escolhas clicando no botão "Create". Depois de criado, o sistema apresenta todas as informações escolhidas, como Network, Backup, retenção de Backup, e o nome da Database. Com poucos cliques, criamos o Autonomous DB, que está pronto para operar.

# **MySQL**

O MySQL é um banco de dados relacional e gerenciado, ideal para aplicações web, aplicações SaaS, IoT e análise de dados em tempo real. A OCI oferece o MySQL HeatWave, uma versão aprimorada do MySQL.

##**Conhecendo MySQL HeatWave**

As principais características do MySQL HeatWave incluem gerenciamento pela OCI, desde atualizações e backups até monitoramentos automáticos.

O MySQL HeatWave é uma engine integrada para acelerar as consultas analíticas do banco de dados, oferecendo segurança avançada e, consequentemente, alto desempenho. Um fato interessante sobre o HeatWave é que, em alguns casos, ele aumentou a velocidade das consultas em até 5.400 vezes - o que é bastante surpreendente.

##**Resolvendo problemas com MySQL HeatWave**

Além dos problemas comuns a todos os tipos de banco de dados que não estão na OCI, um problema específico que queremos destacar é o desempenho lento de bancos de dados para análises. Para solucioná-lo, podemos utilizar o HeatWave.

Em uma solução tradicional, seria necessário configurar um servidor ou um banco de dados separado apenas para esse tipo de análise. No entanto, se já estivermos utilizando o MySQL, podemos migrar ou simplesmente ativar o HeatWave.

##**Exemplificando uso real**

Um exemplo real ocorre quando uma empresa de determinado ramo enfrenta uma demanda repentina de muitos dados para serem analisados. Isso exige alta performance do banco de dados.

Em um ambiente local, ou on-premise, pode ser difícil lidar com essa situação de forma eficiente. Na nuvem, entretanto, podemos migrar ou ativar o serviço do HeatWave, conforme mencionado anteriormente.

##**Listando vantagens do MySQL HeatWave**

O MySQL HeatWave apresenta soluções eficazes, como, por exemplo:

 - Consultas analíticas podem ser até 5.400 vezes mais rápidas;

 - Escalabilidade automática;

 - Gerenciamento totalmente automatizado;

 - Alta disponibilidade e failover automático.

Quando falamos em failover automático, temos a garantia de que os bancos de dados estarão disponíveis 24 horas por dia, 7 dias por semana. Além disso, ao criar o banco de dados, possamos escolher entre uma única instância ou três réplicas, garantindo a alta disponibilidade do serviço.

##**Criando um DB System**

Vamos acessar agora o console da OCI. No menu de navegação na lateral esquerda, vamos acessar até "Databases". Dentro de "Database", podemos clicar tanto em "HeatWave MySQL" diretamente ou em "DBSystems".

Nesse caso, vamos clicar em "DBSystems" para acessar a parte de sistema de banco de dados. Se clicássemos em "HeatWave MySQL", seríamos redirecionados para Overview. Portanto, não há diferença significativa.

Em "DBSystems", basta clicar no botão "Create DB System". Primeiro, devemos escolher dentro os ambientes: Production (produção), Development or testing (desenvolvimento ou teste) e Always Free (sempre gratuito). Vamos selecionar a última opção.

Também podemos definir o compartment, especificar o nome do banco e adicionar uma descrição opcional. Além disso, é possível criar credenciais de administrador, configurando uma senha.

No modo Always Free, não é permitido escolher a alta disponibilidade, que inclui a réplica em três bancos de dados. No entanto, ao mudar para produção, essa opção fica disponível.

Ademais, podemos fazer as configurações de Networking e Placement. Na seção "Configure hardware", chegamos ao ponto principal, que faz a configuração do cluster do HeatWave.

Clicando em "Edit" na configuração do hardware, podemos escolher o shape, ou seja, a quantidade de ECPUs e memória. No modo Free, essas opções não podem ser alteradas.

É possível habilitar ou desabilitar o HeatWave. Com essa opção ativada, podemos escolher o tipo de cluster de HeatWave. Alternativamente, podemos utilizar um MySQL comum. No entanto, se surgir a necessidade de realizar análises complexas, o ideal é habilitar o HeatWave para obter um banco de dados mais rápido.
