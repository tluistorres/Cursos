# **07.4. Encryption**

##**Introduzindo criptografia**

A criptografia  o processo de transformar dados. Isso significa converter dados de texto simples (plaintext) em texto cifrado (ciphertext). Por exemplo, a palavra "transformar" est em texto simples. Quando criptografada, ela se torna um conjunto embaralhado de letras, nmeros e outros caracteres.

O plaintext  o texto compreensvel e visvel, enquanto o ciphertext  o texto cifrado, composto por letras e nmeros aleatrios, e, s vezes, caracteres especficos que no fazem sentido para seres humanos.

Aps a criptografia, temos o processo de descriptografia. Quando o texto se torna ilegvel, precisamos descriptograf-lo. Para isso, pegamos o texto cifrado e utilizamos a chave correta para convert-lo de volta ao texto legvel.

**A chave de criptografia  uma sequncia de letras e nmeros, geralmente armazenada em um arquivo, utilizada por algoritmos criptogrficos para criptografar e descriptografar dados.**

Muitas vezes, falamos sobre o par de chaves, como a chave pblica e a chave privada, que so usadas tanto para criptografia quanto para assinatura digital.

##**Conhecendo tipos de criptografia**

Existem diferentes tipos de criptografia. A criptografia em repouso protege dados armazenados em dispositivos fsicos, bancos de dados ou discos de servidores. Mesmo que um invasor roube o disco, sem a chave de acesso, ele no conseguir ler os dados.

A criptografia em trnsito protege dados em movimento. Um exemplo prtico  o cadeado na URL de um site, indicando o uso do protocolo HTTPS, que garante a criptografia em trnsito. Se um site no tiver o cadeado fechado, ele estar apenas em HTTP, sem a segurana adicional do HTTPS.

Outro tipo de criptografia  a simtrica, na qual usamos uma nica chave para criptografar e descriptografar. Por exemplo, se o usurio A deseja enviar uma mensagem para o usurio B, o usurio A criptografa a mensagem com sua chave, e o usurio B a descriptografa com a mesma chave.

A utilizao da mesma chave em criptografia simtrica apresenta desvantagens, pois ela precisa ser compartilhada entre ambas as partes - o que implica que, em algum momento, essa chave estar em trnsito.

J na criptografia assimtrica, utilizam-se pares de chaves, que  o mtodo mais comum. Nesse caso, temos uma chave pblica e uma chave privada. A chave pblica  utilizada para criptografar, enquanto a chave privada  usada para descriptografar.

Por exemplo, o usurio A gera um par de chaves. O usurio B utiliza a chave pblica, gerada nesse par, para criptografar uma mensagem. Quando a mensagem retorna para o usurio B, ele possui a chave privada. Assim, a chave privada no  compartilhada.

Com isso, no haver problema se essa chave pblica for hackeada ou perdida, pois ela  usada apenas para criptografar.

##**Discutindo sobre algoritmos de criptografia**

Os principais algoritmos de criptografia incluem:

**AES** (*Advanced Encryption Standard*):  simtrico e utiliza chaves de tamanhos como 128, 192 e 256 bits.  muito eficiente e seguro;

**RSA** (*Rivest-Shamir-Adleman*):  assimtrico e utiliza um par de chaves.  mais seguro, porm, mais lento que o AES;

**ECDSA** (*Elliptic Curve Digital Signature Algorithm*): usado para assinatura digital, no para criptografia de dados. Gera chaves menores e seguras com base na criptografia de curvas elpticas.

Alm disso, existe o mdulo de segurana de hardware, conhecido como HSM (Hardware Security Module). Este  um dispositivo fsico que protege e gerencia chaves criptogrficas, realiza operaes de criptografia e assinatura. Assim, ele garante a autenticao forte e resistncia  violao.

Caso algum tente violar o dispositivo, ele apaga as chaves automaticamente.  certificado segundo padres internacionais, como o FIPS 140. O prprio vault da OCI utiliza essa certificao para segurana no ambiente da OCI.

##**Concluso**

Em resumo, a criptografia transforma texto legvel em texto criptografado. Ela pode ser simtrica ou assimtrica, alm de estar em repouso ou em trnsito. Os tipos incluem AES, RSA e o HSM, que  o mdulo de segurana. At o prximo vdeo.

Por sua vez, o objeto criptografado, junto com a chave de dados criptografada, so armazenados no bucket.

Para descriptografar, a OCI envia a chave de dados criptografada para o vault. O vault descriptografa a DEK usando a chave mestra. A DEK  usada para descriptografar o objeto e obter os dados.

Isso  a teoria por trs do vault. No  necessrio saber todos os detalhes para o funcionamento, mas  interessante para a prova de certificao entender esses conceitos de chave de dados DEK e chave mestra.

## **08.1. Pricing**

##**Desafios da precificao na nuvem**

Quando falamos de precificao na nuvem, pode ser um tema complexo e, muitas vezes, em alguns provedores, bastante complicado de lidar. No entanto, na OCI (Oracle Cloud Infrastructure), a abordagem  simples, transparente e acessvel. ##**Modelos de precificao na OCI**

Vamos explorar os modelos de precificao dentro da OCI. Temos o modelo Pay As You Go (pagamento conforme o uso), que significa que pagamos apenas pelos recursos consumidos, sem compromisso inicial ou perodo mnimo de servio. Por exemplo, ao migrar de uma conta gratuita para uma conta paga, pagamos apenas pelo que consumimos, como instncias ou volumes de dados.

A precificao por consumo  diferente do modelo conforme o uso. Ela  aplicada em servios como as Functions. Nesse caso, pagamos apenas no momento exato em que utilizamos o servio, diferente de uma mquina virtual, onde pagamos pelo tempo em que est ativa. Com as Functions, ao executar um comando, pagamos pelo uso; ao parar, cessamos o gasto. Este  um ponto interessante da precificao.

Outro modelo  o dos crditos universais anuais, com o qual estamos mais acostumados a utilizar na OCI. Neste modelo, fechamos um pool de crditos, ou seja, uma coleo de crditos. Por exemplo, podemos fechar um contrato com a OCI de um ano, com um valor de R$100 mil. Teremos R$100 mil em crditos para utilizar ao longo do ano, conforme a necessidade. Um detalhe importante  que, se no utilizarmos todos os R$100 mil em crditos dentro do ano, por exemplo, se sobraram R$20 mil, perderemos esses crditos.

Por fim,  possvel trazer a prpria licena. Essa opo  vantajosa para quem j possui licenas locais (on-premise) e deseja reutiliz-las na nuvem, reduzindo custos com novos contratos.

##**Fatores que impactam a precificao**

Os principais fatores que impactam a precificao so o tamanho do recurso, quanto maior o recurso, maior o custo, e o modelo baseado em consumo. O tipo de recurso tambm interfere: por exemplo, uma mquina virtual tem um custo, um bare metal tem outro, e servios como banco de dados ou HeatWave possuem precificaes especficas.

Alm disso, reutilizar licenas prprias da infraestrutura local ajuda a reduzir custos, enquanto utilizar licenas diretamente na OCI tende a aumentar os gastos. A transferncia de dados  um dos elementos mais relevantes na fatura, podendo representar 25% ou mais do valor total.

##**Vantagens da precificao uniforme e transferncia de dados**

Um ponto interessante sobre a precificao na OCI  que o preo da Oracle  o mesmo em todas as regies do mundo. Por exemplo, se pagarmos R$ 2 hospedado em So Paulo, pagaremos o mesmo valor nos Estados Unidos ou em qualquer outro local da Europa. Isso  vantajoso para a variao de cmbio, pois pagamos o mesmo valor independentemente de onde hospedamos, tornando desnecessrio buscar hospedagem fora do pas. Podemos hospedar com baixa latncia dentro do pas e ter uma precificao uniforme.

Em relao  transferncia de dados, a transferncia entre domnios de disponibilidade, como do Brasil para a Europa, no  cobrada pela OCI. Em outras nuvens, essa transferncia  cobrada, mas na OCI no h essa cobrana. **08.4. Tagging**

Neste vdeo, discutiremos as tags (etiquetas) dentro do ambiente da OCI. As tags so pares chave-valor que podemos usar para organizar melhor nossos recursos na nuvem.

O problema de no utilizarmos esse tipo de recurso  que oodemos ter, por exemplo, dificuldade para encontrar recursos. Quando etiquetamos algo, ou seja, quando adicionamos uma tag, torna-se muito mais simples, prtico e rpido encontrar esses recursos.

```python
Tag 1: enviroment = production
Tag 2: project = alpha
```

Por exemplo, podemos ter a tag 1, que vem com uma chave environment e o valor produo. A tag 2 vem com uma chave project e um valor alfa.

Benefcios das tags
As tags ajudam na organizao de ambientes com muitos recursos, facilitando a identificao por projeto, ambiente ou equipe. Tambm permitem um gerenciamento mais eficiente de custos, possibilitando o monitoramento por equipe ou projeto.

Alm disso, permitem aplicar controle de acesso com base em tags, definindo polticas especficas conforme a equipe ou recurso, mesmo dentro de um nico compartment.

##**Tipos de tags na OCI**

Temos as tags livres, que so simplesmente pares chave-valor. Elas no possuem validao de valores e controle de uso, alm de no exigirem configurao prvia. J as tags definidas so mais avanadas e estruturadas, incluindo um termo adicional: o namespace (espao de nome).

Por exemplo, uma chave livre pode ser environment = production, contendo apenas a chave e o valor. Em contrapartida, uma tag definida incluiria o namespace operations, a chave environment e o valor production. Isso permite especificar valores vlidos de forma mais eficaz e aplicar polticas de acesso, possibilitando um controle mais rigoroso sobre as tags e os recursos.

```python
Livres: environment = production
Definidas: operations.environment = production
```

O funcionamento das tags definidas segue uma estrutura composta por namespace, chave e valor. Por exemplo: namespace operations, chave environment e valor production. Essa organizao define a estrutura completa da tag.

Um ponto importante  que aps criado, o namespace no pode ser excludo, apenas aposentado. Isso garante a preservao do histrico, o que  essencial para auditorias. Mesmo que o namespace project seja aposentado, todo o histrico vinculado a ele e s tags associadas continuar disponvel.
